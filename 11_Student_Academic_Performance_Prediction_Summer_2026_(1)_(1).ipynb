### Question 2: Identify Column Types (2 Marks)
Identify and print the list of numerical and categorical column names in the dataset.

**Expected Output:**
```
Lists of numerical and categorical column names.
```

````python
# Import pandas
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/STUDENTS TIH.csv')

# Get numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Get categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()

# Print the column names
print("Numerical Columns:")
print(numerical_columns)

print("\nCategorical Columns:")
print(categorical_columns)
````

### Question 3: Missing Values (2 Marks)
Check for missing values in the dataset and print the sum of missing values for each column.

**Expected Output:**
```


````python
# Check missing values
missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)
````

### Question 4: Data Imputation (3 Marks)
Fill missing numerical values with the median and missing categorical values with the mode. Print the total number of missing values remaining in the entire dataset.



````python
# Fill numerical columns with median
for col in numerical_columns:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical columns with mode
for col in categorical_columns:
    df[col] = df[col].fillna(df[col].mode()[0])

# Total missing values remaining
print("Total Missing Values:", df.isnull().sum().sum())
````

### Question 5: Label Encoding (3 Marks)
Use `LabelEncoder` from `sklearn.preprocessing` to encode all categorical columns into numerical values. Print the first 5 rows of the updated DataFrame.

**Hint:** Loop through the categorical columns and apply `.fit_transform()`.

**Expected Output:**
```
A pandas DataFrame head with all columns containing numerical values.
```

````python
# Import required libraries
from sklearn.preprocessing import LabelEncoder
# Identify categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Create a LabelEncoder object
le = LabelEncoder()

# Apply Label Encoding to all categorical columns
for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))

# Print the first 5 rows of the updated DataFrame
print("First 5 rows of the encoded DataFrame:")
print(df.head())

````

### Question 6: Feature and Target Separation (3 Marks)
Separate the features `X` and the target `y` (assuming 'performance_category' is the target). Print the shapes of `X` and `y`.

**Expected Output:**
```
Shapes of X and y (e.g., (n_samples, n_features) and (n_samples,)).
```

````python
# Separate features (X) and target (y)
X = df.drop(['performance_category', 'final_exam_score'], axis=1)
y = df['performance_category']

# Print the shapes of X and y
print("Shape of X and Y:", X.shape,y.shape)
````

### Question 7: Train-Test Split (3 Marks)
Split the data into training and testing sets with a test size of 20% and `random_state=42`. Print the shapes of `X_train` and `X_test`.

**Expected Output:**
```
Shapes of X_train and X_test.
```

````python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)

# Print the shapes of X_train and X_test
print("Shape of X_train and X_test:", X_train.shape, X_test.shape)
````

### Question 8: Feature Scaling (3 Marks)
Scale the features using `MinMaxScaler` from `sklearn.preprocessing`. Fit on `X_train` and transform both `X_train` and `X_test`. Print the minimum and maximum values of the scaled `X_train`.



````python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit on X_train and transform X_train and X_test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Print the minimum and maximum values of the scaled X_train
print("Minimum value of X_train_scaled:", X_train_scaled.min())
print("Maximum value of X_train_scaled:", X_train_scaled.max())
````

### Question 9: Decision Tree Classifier (5 Marks)
Initialize and train a `DecisionTreeClassifier` with `random_state=42`. Make predictions on the test set and print the accuracy score.

**Expected Output:**
```
A numerical value representing the accuracy score.
```

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Train the model
dt_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = dt_classifier.predict(X_test_scaled)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy Score:", accuracy)
````
